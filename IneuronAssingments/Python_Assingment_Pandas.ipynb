```
import pandas as pd
import numpy as np
```

```
#Data Fram Creation Using The Avalable Data

df = pd.DataFrame({
'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],

'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],

'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],

'Airline': ['KLM(!)', ' (12)', '(British Airways. )',

'12. Air France', '"Swiss Air"']}) 
```

```
df  #Displyaing A Dataset
```

```
df.loc[df['FlightNumber'].isna(),'FlightNumber'] #Finding a NAN Values
```

```
df.loc[df[df['FlightNumber'].isna()].index -1,'FlightNumber'].astype(int)+10  #Adding a Values +10 
```

```
#First Way

for i in range(1, df['FlightNumber'].count() + 1):
      if pd.isnull(df.loc[i,'FlightNumber']):
          df.loc[i, 'FlightNumber'] = df.loc[i-1, 'FlightNumber'] + 10

df['FlightNumber'] = df['FlightNumber'].astype(int)           
```

```
df #Displaying a values
```

```
#Second Way


df['FlightNumber'][1] = 10055
df['FlightNumber'][3] = 10075
df['FlightNumber'] = df['FlightNumber'].astype(int)
df
```
